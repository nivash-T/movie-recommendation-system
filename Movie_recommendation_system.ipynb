```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data=pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
#extract necessary feature
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
```

```python
# Replace missing values with an empty string for each selected feature
for feature in selected_features:
    movies_data[feature].fillna('', inplace=True)
```

```python
#combining features
combine_features=movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combine_features)
```

```python
#converting text data to feature vector
vectorizer=TfidfVectorizer()
feature_vector=vectorizer.fit_transform(combine_features)
print(feature_vector)
```

```python
# verifying similarity using cosine similarity
similarity=cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
# getting movie name
movie_name=input('enter your favourite movie name : ')
```

```python
# craeting seperate list for titles
list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding close match using similarity score
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
# taking closest match score as movie_name(variable) score
closest_match=find_close_match[0]
```

```python
# find index of closeast match
index_of_movie=movies_data[movies_data.title==closest_match]['index'].values[0]
```

```python
print(index_of_movie)
```

```python
# getting list of similarity score for  movies
similarity_score=list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
# sorting in decending order of similarity scores
sorted_similar_movies= sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#picking up top highest similarity scores movies
print('movie suggested for u : \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if i<=20:
    print(i,'.',title_from_index)
    i=i+1
```
